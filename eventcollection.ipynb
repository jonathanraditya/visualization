# EventCollection Demo
Plot two curves, then use `EventCollection`s to mark the locations of the x and y data points on the respective axes for each curve.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.collections import EventCollection as EC

!python --version
print(f'Matplotlib version: {mpl.__version__}')
print(f'Numpy version: {np.__version__}')
```

```python
np.random.seed(0)
x1 = np.sort(np.random.random(20))
x2 = np.sort(np.random.random(7))
y1 = np.exp(x1)
y2 = np.exp(x2)
y2 = y2.max() - y2

fig, ax = plt.subplots()
ax.plot(x1, y1, label=2)
ax.plot(x2, y2, label=1)

ax.set(title='line plot with data points', xlim=(0,1), ylim=(0,3))
ax.legend()

xe1 = EC(x1, color='tab:blue', linelength=0.5, orientation='horizontal')
xe2 = EC(x2, color='tab:orange', linelength=0.5, orientation='horizontal')
ye1 = EC(y1, color='tab:blue', linelength=0.05, orientation='vertical')
ye2 = EC(y2, color='tab:orange', linelength=0.05, orientation='vertical')

for i in (xe1, xe2, ye1, ye2): ax.add_collection(i)

plt.show()
```
