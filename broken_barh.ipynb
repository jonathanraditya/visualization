# Broken Barh
Make a "broken" horizontal bar plot, i.e., one with gaps

```python
import matplotlib.pyplot as plt
import matplotlib as mpl

!python --version
print(f'Matplotlib version: {mpl.__version__}')
```

```python
fig, ax = plt.subplots()

ax.broken_barh(((110, 20), (150,10)), (10, 9), facecolor='tab:blue')
ax.broken_barh(((15,50), (100,20), (130,10)), (20,9), facecolor=('tab:orange', 'tab:green', 'tab:red'))

ax.set_xlabel('seconds since start')
ax.set_xlim((0,200))
ax.set_yticks((15,25), ('Bill', 'Jim'))
ax.set_ylim((5,35))
ax.grid(True)

ax.annotate('race interrupted', (61,26), 
            xytext=(0.8,0.9), textcoords='axes fraction', 
            arrowprops=dict(facecolor='black', shrink=0.05), 
            fontsize=16, 
            horizontalalignment='right', verticalalignment='top')

plt.show()
```
