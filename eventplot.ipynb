# Eventplot demo
An `eventplot` showing sequences of events with various line properties. The plot is shown in both horizontal and vertical orientations

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

!python --version
print(f'Numpy version: {np.__version__}')
print(f'Matplotlib version: {mpl.__version__ }')
```

```python
v = 3
n = 10
d1 = np.random.random((v,n))
d2 = np.random.gamma(v, size=(v*n, v*n))
c = [f'C{i}' for i in range(v)]
o = np.random.random(v) * v
o = np.sort(o)
l = o * v / n

fig, ax = plt.subplots(2,2)

ax[0,0].eventplot(d1, colors=c, lineoffsets=o, linelengths=l)
ax[1,0].eventplot(d1, colors=c, lineoffsets=o, linelengths=l, orientation='vertical')

ax[0,1].eventplot(d2, colors='black')
ax[1,1].eventplot(d2, colors='black', orientation='vertical')

plt.show()
```
