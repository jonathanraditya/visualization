# Errorbar limit selection
Illustration of selectively drawing lower and/or upper limit symbols on errorbars using the parameters `uplims`, `lolims` of `errorbar`.

Alternatively, you can use 2xN values to draw errorbars in only one direction.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

!python --version
print(f'Matplotlib version: {mpl.__version__}')
print(f'Numpy version: {np.__version__}')
```

```python
n = 10
x = np.arange(n)
y = np.sin(x / (2 * n) * np.pi)


fig, ax = plt.subplots()
ax.errorbar(x, y, np.linspace(0,0.1, n), uplims=[True, False] * int(n / 2), lolims=True)

plt.show()
```

# Errorbar subsampling
The parameter *errorevery* of `Axes.errorbar` can be used to draw error bars only on a subset of data points. This is particularly useful if there are many data points with similar errors.

```python
n = 20
x = np.linspace(0,4,n)
y1 = np.sin(x / n * np.pi * 2 + 0.05)
y2 = np.sin(x / n * np.pi * 2 + 0.5)
err = np.linspace(0,0.1, n)

fig, ax = plt.subplots(1, 3, sharey=True, sharex=True, layout='constrained', figsize=(9,3))

ax[0].set_title('all errorbars')
ax[0].errorbar(x, y1, yerr=err)
ax[0].errorbar(x, y2, yerr=err)

ax[1].set_title('only every 6th errorbar')
ax[1].errorbar(x, y1, yerr=err, errorevery=6)
ax[1].errorbar(x, y2, yerr=err, errorevery=6)

ax[2].set_title('second series shifted by 3')
ax[2].errorbar(x, y1, yerr=err, errorevery=(0,6))
ax[2].errorbar(x, y2, yerr=err, errorevery=(3,6))

fig.suptitle('Errorbar subsampling')
plt.show()
```
