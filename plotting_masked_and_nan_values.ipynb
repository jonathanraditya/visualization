# Plotting masked and NaN values

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

!python --version
print(f'Matplotlib version: {mpl.__version__}')
print(f'Numpy version: {np.__version__}')
```

```python
x = np.linspace(-np.pi/2, np.pi/2, 41)
y = np.cos(x) ** 3
ylim = 0.7

# Points removed
x2 = x[y < ylim]
y2 = y[y < ylim]

# Masked values
y3 = np.ma.masked_where(y > ylim, y)

# NaN values
y4 = y.copy()
y4[y > ylim] = np.nan

fig, ax = plt.subplots()

ax.plot(x * 0.1, y, 'o', color='gray', label='No mask')
ax.plot(x2 * 0.25, y2, 'o', color='tab:blue', label='Points removed')
ax.plot(x * 0.5, y3, 'o', color='tab:orange', label='Masked values')
ax.plot(x * 0.75, y4, 'o', color='tab:green', label='NaN values')

ax.set(title='Masked and NaN data')
ax.legend()



plt.show()
```
